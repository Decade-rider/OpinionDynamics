```python
def add_col(M):
    if len(M.shape) >1:
        M_new = np.zeros([M.shape[0], M.shape[1]+1])
        M_new[:, :-1] = M
    else:
        M_new = np.zeros(M.shape[0]+1)
        M_new[:-1] = M
    return M_new

def next_step(vec, agents):
    row_matrix = np.tile(vec, (agents,1))
    col_matrix = row_matrix.transpose()
    dist_matrix = row_matrix - col_matrix
    tf_1 = dist_matrix <= e_r
    tf_2 = dist_matrix >= -e_l
    considered = np.logical_and(tf_1, tf_2)
    considered_ops = np.ma.masked_where(~considered, row_matrix)
    return np.mean(considered_ops, axis = 1)

def dynamics(x_0, e_l, e_r):
    times = np.array([0,1], dtype=int)
    agents = x_0.shape[0]
    result = np.zeros([agents, 2])
    result[:,0] = x_0
    
    control = 10**(-5)
    
    # do first iteration    
    result[:, 1] = next_step(result[:,0], agents)
        
    t = 2
    #while np.any(abs(result[:,t-2] - result[:,t-1]) > control):
    for i in range(8):
        times = add_col(times)
        times[-1] = t
        result = add_col(result)
        result[:, t] = next_step(result[:, t-1], agents)
        t += 1
    return result
    
def display(x_0, e_l, e_r):
    result = dynamics(x_0, e_l, e_r)
    times = np.linspace(0, result.shape[1], result.shape[1])
    agents = result.shape[0]
    cmap = plt.get_cmap('viridis')
    colors = [cmap(i) for i in np.linspace(0, 1, agents)]
    for a in range(agents):
        plt.plot(times, result[a, :], color=colors[a], linewidth=0.7)
        
np.random.seed(1)
x_0 = np.sort(np.random.uniform(0,1,50))

fig, ax = plt.subplots(2, 2)
fig.set_size_inches(10, 8)


plt.sca(ax[0, 0])
e_l = 0.05
e_r = 0.05
display(x_0, e_l, e_r)
plt.xlabel("Time")
plt.ylabel("Opinion")
plt.title('A ($\epsilon = 0.05$)', loc='left')


plt.sca(ax[0, 1])
e_l = 0.055
e_r = 0.055
display(x_0, e_l, e_r)
plt.xlabel("Time")
plt.ylabel("Opinion")
plt.title('B ($\epsilon = 0.055$)', loc='left')

plt.sca(ax[1, 0])
e_l = 0.26
e_r = 0.26
display(x_0, e_l, e_r)
plt.xlabel("Time")
plt.ylabel("Opinion")
plt.title('C ($\epsilon = 0.26$)', loc='left')

plt.sca(ax[1,1])
e_l = 0.3
e_r = 0.3
display(x_0, e_l, e_r)
plt.xlabel("Time")
plt.ylabel("Opinion")
plt.title('D ($\epsilon = 0.3$)', loc='left')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def next_step(vec, agents):
    row_matrix = np.tile(vec, (agents,1))
    col_matrix = row_matrix.transpose()
    dist_matrix = row_matrix - col_matrix
    tf_1 = dist_matrix <= e_r
    tf_2 = dist_matrix >= -e_l
    considered = np.logical_and(tf_1, tf_2)
    considered_ops = np.ma.masked_where(~considered, row_matrix)
    weights = 1/np.ma.count(considered_ops, axis=1)
    weightmat = np.tile(weights, (agents,1)).transpose()
    interactMat = np.ma.masked_where(~considered, weightmat).filled(0)
    #print(interactMat)
    return np.mean(considered_ops, axis = 1), np.sum(interactMat**2), np.trace(interactMat)

def dynamics(x_0, e_l, e_r):
    times = np.array([0,1], dtype=int)
    agents = x_0.shape[0]
    result = np.zeros([agents, 2])
    result[:,0] = x_0
    
    control = 10**(-5)
    
    asqs = []
    trs = []
    # do first iteration
    nextstep = next_step(result[:,0], agents)
    result[:, 1] = nextstep[0]
    asqs.append(nextstep[1])
    trs.append(nextstep[2])
        
    t = 2
    while np.any(abs(result[:,t-2] - result[:,t-1]) > control):
        times = add_col(times)
        times[-1] = t
        result = add_col(result)
        nextstep = next_step(result[:, t-1], agents)
        result[:, t] = nextstep[0]
        asqs.append(nextstep[1])
        trs.append(nextstep[2])
        t += 1
    return result, asqs, trs

def cluster_analysis(x_0, e_l, e_r):
    res = dynamics(x_0, e_l, e_r)
    dyn = res[0]
    fin_interMat = res[1]
    stable_pattern = dyn[:, -1]
    t = dyn.shape[1]
    
    opinions = [stable_pattern[0]]
    count_in_cluster = [1]
    stability_control = min(e_l, e_r)/4
    k = 0
    for i in range(1, stable_pattern.shape[0]):
        if abs(stable_pattern[i-1] - stable_pattern[i]) < stability_control:
            count_in_cluster[k] +=1
        else:
            opinions.append(stable_pattern[i])
            count_in_cluster.append(1)
            k += 1
    return opinions, count_in_cluster, t

def setup_arrays(samples, esteps, es):
    real = np.zeros([samples, esteps])
    trace = np.zeros([samples, esteps])
    conj_2R = np.zeros(esteps)
    for k in range(esteps):
        conj_2R[k] = min(agents, 1/(2*es[k]))
    return real, trace, conj_2R

def get_initialCondition(agents, seed):
    np.random.seed(seed)
    return np.sort(np.random.uniform(0,1,agents))

def get_clusterResults(x_0, e_l, e_r, agents):
    num = len(cluster_analysis(x_0, e_l, e_r)[0])
    tr = next_step(x_0, agents)[2]
    return num, tr

def plot_clusterResult(es, conj_2R, trace, real):
    plt.plot(es, conj_2R, label="2R-conj.")
    plt.plot(es, np.mean(trace, axis = 0), label="trace")
    plt.plot(es, np.mean(real, axis = 0), label="num. simul.")
    plt.xlabel("$\epsilon$")
    plt.ylabel("number of clusters")
    plt.legend()


samples = 100
esteps = 20

fig, ax = plt.subplots(3, 3)
fig.set_size_inches(10, 10)

    
agentss = [10, 50, 100]
headings = ['A', 'B', 'C']
for a in range(3):
    agents = agentss[a]
    
    plt.sca(ax[a, 0])
    es = np.linspace(0.0001,0.05, esteps)
    real, trace, conj_2R = setup_arrays(samples, esteps, es)
    for i in range(samples):
        x_0 = get_initialCondition(agents, i)
        for j in range(esteps):
            e_l = es[j]
            e_r = e_l
            real[i,j], trace[i,j] = get_clusterResults(x_0, e_l, e_r, agents)
    plot_clusterResult(es, conj_2R, trace, real)
    plt.title(headings[a] + ' ($n = $' + str(agents) + ')', loc='left')



    plt.sca(ax[a, 1])
    es = np.linspace(0.05,0.2, esteps)
    real, trace, conj_2R = setup_arrays(samples, esteps, es)
    for i in range(samples):
        x_0 = get_initialCondition(agents, i)
        for j in range(esteps):
            e_l = es[j]
            e_r = e_l
            real[i,j], trace[i,j] = get_clusterResults(x_0, e_l, e_r, agents)
    plot_clusterResult(es, conj_2R, trace, real)



    plt.sca(ax[a, 2])
    es = np.linspace(0.2,0.7, esteps)
    real, trace, conj_2R = setup_arrays(samples, esteps, es)
    for i in range(samples):
        x_0 = get_initialCondition(agents, i)
        for j in range(esteps):
            e_l = es[j]
            e_r = e_l
            real[i,j], trace[i,j] = get_clusterResults(x_0, e_l, e_r, agents)
    plot_clusterResult(es, conj_2R, trace, real)
    
plt.tight_layout()
plt.show()
```

```python
#HK_lr_slice
def next_step_FixAs(vec, agents, e_l, e_r):
    row_matrix = np.tile(vec, (agents,1))
    col_matrix = row_matrix.transpose()
    dist_matrix = row_matrix - col_matrix
    tf_1 = dist_matrix <= e_r
    tf_2 = dist_matrix >= -e_l
    considered = np.logical_and(tf_1, tf_2)
    considered_ops = np.ma.masked_where(~considered, row_matrix)
    weights = 1/np.ma.count(considered_ops, axis=1)
    weightmat = np.tile(weights, (agents,1)).transpose()
    interactMat = np.ma.masked_where(~considered, weightmat).filled(0)
    return np.mean(considered_ops, axis = 1), np.trace(interactMat)

def calculate_trace(samples, fixed_er, N, agents, e_ls):
    # calculate trace
    trace = np.zeros([samples, N])
    for i in range(samples):
        np.random.seed(i)
        x_0 = np.sort(np.random.uniform(0,1,agents))
        for j in range(N):
            e_l = e_ls[j]
            e_r = fixed_er
            trace[i,j] = next_step_FixAs(x_0, agents, e_l, e_r)[1]
    return trace

def plot_numberClusters(e_ls, pos, countgrid_2R, trace, countgrid, fixed_er):
    ylims = [5, 50]
    plt.plot(e_ls, countgrid_2R[:, pos], label="$1/(2\max(\epsilon_l, \epsilon_r))$")
    plt.plot(e_ls, np.mean(trace, axis = 0), label="trace")
    plt.plot(e_ls, countgrid[:, pos], label="num. simul.")
    plt.ylim(ylims)
    plt.vlines(fixed_er, ylims[0], ylims[1], color="red", linestyle = "--")
    plt.xlabel("$\epsilon_l$")
    plt.ylabel("number of clusters")
    plt.legend()


e_l_range = [0.001,0.05]
e_r_range = [0.001,0.05]
N = 50
agents = 50
samples = 100 # for calculating trace
e_ls = np.linspace(e_l_range[0], e_l_range[1], N)
e_rs = np.linspace(e_r_range[0], e_r_range[1], N)
countgrid = np.genfromtxt('./savedResults/clusters_HK_FixAs.csv', delimiter=';')

countgrid_2R = np.zeros([N, N])

# calculate 1/(2* max(e_l, e_r)) for every e_l e_r combination
for i in range(N):
    e_l = e_ls[i]
    for j in range(N):
        e_r = e_rs[j]
        res = 1/(2*max(e_l,e_r))
        if res > 50:
            countgrid_2R[i,j] = 50
        else:
            countgrid_2R[i,j] = res
            


fig, ax = plt.subplots(1, 3)
fig.set_size_inches(12, 4)

plt.sca(ax[0])
pos = 10
fixed_er_1 = e_rs[pos]
trace = calculate_trace(samples, fixed_er_1, N, agents, e_ls)
plot_numberClusters(e_ls, pos, countgrid_2R, trace, countgrid, fixed_er_1)
plt.title("A ($\epsilon_r = $" + str(fixed_er_1) + ")", loc="left")

plt.sca(ax[1])
pos = 20
fixed_er_2 = e_rs[pos]
trace = calculate_trace(samples, fixed_er_2, N, agents, e_ls)
plot_numberClusters(e_ls, pos, countgrid_2R, trace, countgrid, fixed_er_2)
plt.title("B ($\epsilon_r = $" + str(fixed_er_2) + ")", loc="left")

plt.sca(ax[2])
pos = 40
fixed_er_3 = e_rs[pos]
trace = calculate_trace(samples, fixed_er_3, N, agents, e_ls)
plot_numberClusters(e_ls, pos, countgrid_2R, trace, countgrid, fixed_er_3)
plt.title("C ($\epsilon_r = $" + str(fixed_er_3) + ")", loc="left")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

e_range = [0.001,0.5]
m_range = [0,1]
N = 100
sample = 100
agents = 50
es = np.linspace(e_range[0], e_range[1], N)
ms = np.linspace(m_range[0], m_range[1], N)
countgrid_final_as = np.genfromtxt('./savedResults/clusters_HK_OpdepAs.csv', delimiter=';')

def next_step_OpdepAs(vec, agents, m, e):
    const = (1-m)/2
    row_matrix = np.tile(vec, (agents,1))
    col_matrix = row_matrix.transpose()
    dist_matrix = row_matrix - col_matrix
    b_r = m * col_matrix[:,0] + const*np.ones(agents)
    b_l = np.ones(agents) - b_r
    e_r = np.zeros([agents,1])
    e_r[:,0] = e*b_r
    e_l = np.zeros([agents,1])
    e_l[:,0] = e*b_l
    tf_1 = dist_matrix <= e_r
    tf_2 = dist_matrix >= -e_l
    considered = np.logical_and(tf_1, tf_2)
    considered_ops = np.ma.masked_where(~considered, row_matrix)
    weights = 1/np.ma.count(considered_ops, axis = 1)
    weightmat = np.tile(weights, (agents, 1)).transpose()
    interactMat = np.ma.masked_where(~considered, weightmat).filled(0)
    return np.mean(considered_ops, axis = 1), np.trace(interactMat)

def clusters(pos):
    m = ms[pos]
    counts = countgrid_final_as[:, pos]
    conj2 = 1/(es * (m+1))
    for k in range(N):
        if conj2[k] >agents:
            conj2[k] = agents
        if conj2[k] >agents:
            conj2[k] = agents
    count_tr = np.zeros([sample, N])
    for i in range(sample):
        np.random.seed(i)
        x_0 = np.sort(np.random.uniform(0,1,agents))
        for j in range(N):
            e = es[j]
            ns = next_step_OpdepAs(x_0, agents, m, e)[1]
            count_tr[i,j] = ns
    avg_count_tr = np.mean(count_tr, axis = 0)
    return conj2, avg_count_tr, counts

def plot_result(conj2, avg_count_tr, counts, startPos, endPos):
    plt.plot(es[startPos:endPos], conj2[startPos:endPos], label="$f(\\bar{\epsilon})$")
    plt.plot(es[startPos:endPos], avg_count_tr[startPos:endPos], label = "trace")
    plt.plot(es[startPos:endPos], counts[startPos:endPos], label="num. simul.")
    plt.legend()
    plt.ylabel("Number of clusters")
    plt.xlabel("$\\bar{\epsilon}$")

    
fig, ax = plt.subplots(3, 2)
fig.set_size_inches(10, 9)

plt.sca(ax[0,0])
p = 10
cl = clusters(p)
plot_result(cl[0], cl[1], cl[2], 0, 50)
plt.title("A ($m = $" + str(np.round(ms[p], 1)) + ")", loc = "left")

plt.sca(ax[0,1])
plot_result(cl[0], cl[1], cl[2], 50, 100)

plt.sca(ax[1,0])
p = 20
cl = clusters(p)
plot_result(cl[0], cl[1], cl[2], 0, 50)
plt.title("B ($m = $" + str(np.round(ms[p], 1)) + ")", loc = "left")

plt.sca(ax[1,1])
plot_result(cl[0], cl[1], cl[2], 50, 100)

plt.sca(ax[2,0])
p = 50
cl = clusters(p)
plot_result(cl[0], cl[1], cl[2], 0, 50)
plt.title("C ($m = $" + str(np.round(ms[p], 1)) + ")", loc = "left")

plt.sca(ax[2,1])
plot_result(cl[0], cl[1], cl[2], 50, 100)

plt.tight_layout()
plt.show()
```
